## Now let's merge the glass data for K24 SWRZ first, then Ulu data (1974) from Wieser et al. 

### First import the glass EDS from June 2024 eruption

```python
KD24GlassEDS=pd.read_excel(MasterFolder+slash+'Data'+slash+"Unpubd data from Kendra"+slash+'DeVitre_samplesdata_01312025.xlsx',sheet_name='EDXRF',skiprows=7)
KD24GlassEDS=KD24GlassEDS.loc[:, ~KD24GlassEDS.columns.str.contains('Unnamed')]
KD24GlassEDS['Sample'] = ['KD24' for sample_name in KD24GlassEDS['SampleName']]
KD24GlassEDS.insert(1, 'Sample', KD24GlassEDS.pop('Sample'))

KD24GlassEDS
```

### Average the Glass data

```python
grouped = KD24GlassEDS.groupby('Sample')

# Separate numeric and non-numeric columns
numeric_cols = KD24GlassEDS.select_dtypes(include='number')
non_numeric_cols = KD24GlassEDS.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_KD24GlassEDS_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_KD24GlassEDS_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_KD24GlassEDS_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate KD24GlassEDS_averageds and reindex columns to the original order
KD24GlassEDS_averaged = pd.concat([numeric_KD24GlassEDS_averaged_mean, non_numeric_KD24GlassEDS_averaged], axis=1)
KD24GlassEDS_averaged = KD24GlassEDS_averaged.reindex(columns=KD24GlassEDS.columns)

# Add standard deviation columns after reindexing
KD24GlassEDS_averaged = pd.concat([KD24GlassEDS_averaged, numeric_KD24GlassEDS_averaged_std.add_suffix('_STD')], axis=1)

# Reset index to get a DataFrame
KD24GlassEDS_averaged = KD24GlassEDS_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
KD24GlassEDS_averaged = pd.merge(KD24GlassEDS_averaged, row_counts[['Sample', 'averaged?']], on='Sample', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
KD24GlassEDS_averaged['averaged?'] = KD24GlassEDS_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# KD24GlassEDS_averaged=KD24GlassEDS_averaged[~(KD24GlassEDS_averaged['filename'].str.contains("186-9-118-A"))]
KD24GlassEDS_averaged=KD24GlassEDS_averaged.drop(columns={'SampleName'})
KD24GlassEDS_averaged_renamed = KD24GlassEDS_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample' else 'Sample')

KD24GlassEDS_averaged_renamed.to_excel(compilation_folder+'/'+"SWRZ_GlassEDS_averaged.xlsx")


KD24GlassEDS_averaged_renamed
```

### Combine the averaged glasses together

```python
averaged_glasses=KD24GlassEDS_averaged_renamed.copy()
averaged_glasses
```

## Now merge and export the merged data

```python
combined_df['Sample'] = combined_df['Sample'].astype(str)
combined_df_wglass = combined_df.merge(averaged_glasses, on='Sample', how='left')
# combined_df_wglass.to_clipboard(excel=True)

combined_df_wglass.to_excel(compilation_folder+slash+'KD24_Raman_EDS_Gl_Merged.xlsx')

combined_df_wglass
```

# Now let's merge the EBSD

```python
EBSD_data=pd.read_excel(MasterFolder +slash+"Data"+slash+'Data_KS24-628'+slash+'EBSD'+slash+'Combined_EBSD'+slash+'EBSD_combined_all.xlsx')
EBSD_data=EBSD_data.rename(columns={'EBSD_name':'EBSD_simple_name','grainID':'EBSD_grainID','Filename':'EBSD_filename'})

final_df=pd.merge(left=combined_df_wglass,on=['EBSD_simple_name','EBSD_grainID'],right=EBSD_data,suffixes=('', '_EBSDdata'),how='left')

##### Export fully merge data ####

final_df.to_excel(compilation_folder+slash+"KS24-628_Final_data_wEBSD.xlsx")
final_df
```
