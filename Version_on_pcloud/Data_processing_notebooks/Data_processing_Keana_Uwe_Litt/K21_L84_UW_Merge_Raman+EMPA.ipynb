## Now let's merge the glass data (SAVE THIS FOR LATER, I DON'T HAVE IT)

### First import the glass data for each eruption

```python
Ulu_all=pd.read_excel(MasterFolder+slash+"Data"+slash+"Data_SWRZ"+slash+"1974_olivine_glass_data"+slash+'UluData.xlsx',
                         sheet_name='Matrix Glass').dropna(axis=0,how='all')
Ulu_all['Eruption']=Ulu_all['Eruption'].astype('str')
Ulu_919=Ulu_all[Ulu_all['Eruption'].str.contains('919')]\

Ulu_919_filt=Ulu_919.loc[:,(~Ulu_919.columns.str.contains('ppm|corrected|Duration|Spot|Batch'))]
Ulu_919_filt.columns = Ulu_919_filt.columns.str.replace('_Liq', '')
Ulu_919_filt=Ulu_919_filt.rename(columns={'Eruption':'Sample','Glass Fragment ID':'Sample Name','Total':'Total_wt%','Mg Number':'Mg#'})
Ulu_919_filt['Sample'] = Ulu_919_filt['Sample'].astype(str).str.split('.').str[0]
Ulu_919_filt
```

### Now average the Ulu data 

```python
grouped = Ulu_919_filt.groupby('Sample')


# Separate numeric and non-numeric columns
numeric_cols = Ulu_919_filt.select_dtypes(include='number')
non_numeric_cols = Ulu_919_filt.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_Ulu_Glass_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_Ulu_Glass_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_Ulu_Glass_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate Ulu_Glass_averageds and reindex columns to the original order
Ulu_Glass_averaged = pd.concat([numeric_Ulu_Glass_averaged_mean, non_numeric_Ulu_Glass_averaged], axis=1)
Ulu_Glass_averaged = Ulu_Glass_averaged.reindex(columns=Ulu_919_filt.columns)

# Add standard deviation columns after reindexing
Ulu_Glass_averaged = pd.concat([Ulu_Glass_averaged, numeric_Ulu_Glass_averaged_std.add_suffix('_STD')], axis=1)

# Reset index to get a DataFrame
Ulu_Glass_averaged = Ulu_Glass_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
Ulu_Glass_averaged = pd.merge(Ulu_Glass_averaged, row_counts[['Sample', 'averaged?']], on='Sample', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
Ulu_Glass_averaged['averaged?'] = Ulu_Glass_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# Ulu_Glass_averaged=Ulu_Glass_averaged[~(Ulu_Glass_averaged['filename'].str.contains("186-9-118-A"))]
Ulu_Glass_averaged=Ulu_Glass_averaged.drop(columns={'Sample Name'})

Ulu_Glass_averaged_renamed = Ulu_Glass_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample' else 'Sample')

Ulu_Glass_averaged_renamed.to_excel(compilation_folder+'/'+"SWRZ_UluGlass_averaged.xlsx")


Ulu_Glass_averaged_renamed
```
