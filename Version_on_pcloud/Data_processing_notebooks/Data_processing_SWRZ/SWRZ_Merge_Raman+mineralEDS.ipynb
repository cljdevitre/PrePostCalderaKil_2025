## Now let's merge the glass data for K24 SWRZ first, then Ulu data (1974) from Wieser et al. 

### First import the glass EDS from June 2024 eruption

```python
K24GlassEDS=pd.read_excel(compilation_folder+slash+'GlassData2024.xlsx')
K24GlassEDS=K24GlassEDS.loc[:, ~K24GlassEDS.columns.str.contains('Unnamed')]
K24GlassEDS['Sample_short'] = ['JDFD2' if 'JDF' in sample_name else 'K24' for sample_name in K24GlassEDS['Sample Name']]
K24GlassEDS.insert(1, 'Sample_short', K24GlassEDS.pop('Sample_short'))

K24GlassEDS
```

### Average the Glass data

```python
grouped = K24GlassEDS.groupby('Sample_short')

# Separate numeric and non-numeric columns
numeric_cols = K24GlassEDS.select_dtypes(include='number')
non_numeric_cols = K24GlassEDS.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_K24GlassEDS_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_K24GlassEDS_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_K24GlassEDS_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate K24GlassEDS_averageds and reindex columns to the original order
K24GlassEDS_averaged = pd.concat([numeric_K24GlassEDS_averaged_mean, non_numeric_K24GlassEDS_averaged], axis=1)
K24GlassEDS_averaged = K24GlassEDS_averaged.reindex(columns=K24GlassEDS.columns)

# Add standard deviation columns after reindexing
K24GlassEDS_averaged = pd.concat([K24GlassEDS_averaged, numeric_K24GlassEDS_averaged_std.add_suffix('_STD')], axis=1)

# Reset index to get a DataFrame
K24GlassEDS_averaged = K24GlassEDS_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
K24GlassEDS_averaged = pd.merge(K24GlassEDS_averaged, row_counts[['Sample_short', 'averaged?']], on='Sample_short', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
K24GlassEDS_averaged['averaged?'] = K24GlassEDS_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# K24GlassEDS_averaged=K24GlassEDS_averaged[~(K24GlassEDS_averaged['filename'].str.contains("186-9-118-A"))]
K24GlassEDS_averaged.to_excel(compilation_folder+'/'+"SWRZ_GlassEDS_averaged.xlsx")

K24GlassEDS_averaged
```

### Now merge the average glass composition for K24 to the main frame

```python
K24GlassEDS_averaged_renamed = K24GlassEDS_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample_short' else 'Sample')

combined_df_wglass = combined_df.merge(K24GlassEDS_averaged_renamed, on='Sample', how='left')
combined_df_wglass
```

### Now let's get the Ulu data (1974)

```python
Ulu_MG_input=pt.import_excel(MasterFolder+slash+"Data"+slash+"Data_SWRZ"+slash+"1974_olivine_glass_data"+slash+'UluData.xlsx',
                         sheet_name='Matrix Glass')
Ulu_all=Ulu_MG_input['my_input'] ## All columns
Ulu_Liqs_all=Ulu_MG_input['Liqs'] ## Just Liquid columns
Ulu_919=Ulu_Liqs_all.loc[Ulu_all['Eruption']==919].copy()

Ulu_919['Sample_ID_Liq'] = Ulu_919['Sample_ID_Liq'].astype(str)
Ulu_919['Mg#']=pt.calculate_liq_mgno(liq_comps=Ulu_919, Fe3Fet_Liq=0.18)
Ulu_919.columns = Ulu_919.columns.str.replace('_Liq', '')

Ulu_919.loc[:,'Sample_ID']='919'
Ulu_919.head()
```

### Now average the Ulu data 

```python
grouped = Ulu_919.groupby('Sample_ID')

# Separate numeric and non-numeric columns
numeric_cols = Ulu_919.select_dtypes(include='number')
non_numeric_cols = Ulu_919.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_Ulu_Glass_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_Ulu_Glass_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_Ulu_Glass_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate Ulu_Glass_averageds and reindex columns to the original order
Ulu_Glass_averaged = pd.concat([numeric_Ulu_Glass_averaged_mean, non_numeric_Ulu_Glass_averaged], axis=1)
Ulu_Glass_averaged = Ulu_Glass_averaged.reindex(columns=Ulu_919.columns)

# Add standard deviation columns after reindexing
Ulu_Glass_averaged = pd.concat([Ulu_Glass_averaged, numeric_Ulu_Glass_averaged_std.add_suffix('_STD')], axis=1)

# Reset index to get a DataFrame
Ulu_Glass_averaged = Ulu_Glass_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
Ulu_Glass_averaged = pd.merge(Ulu_Glass_averaged, row_counts[['Sample_ID', 'averaged?']], on='Sample_ID', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
Ulu_Glass_averaged['averaged?'] = Ulu_Glass_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# Ulu_Glass_averaged=Ulu_Glass_averaged[~(Ulu_Glass_averaged['filename'].str.contains("186-9-118-A"))]
Ulu_Glass_averaged.to_excel(compilation_folder+'/'+"SWRZ_UluGlass_averaged.xlsx")

Ulu_Glass_averaged
```

## Now export the merged data

```python

Ulu_Glass_averaged_renamed = Ulu_Glass_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample_ID' else 'Sample')
Ulu_Glass_averaged_renamed['Sample'] = Ulu_Glass_averaged_renamed['Sample'].astype('float64')
combined_df_wallglass = combined_df_wglass.merge(Ulu_Glass_averaged_renamed, on='Sample', how='left')
combined_df_wallglass.to_excel(compilation_folder+slash+'SWRZ_Raman_EDS_Gl_Merged.xlsx')

combined_df_wallglass
```

```python
# Rename columns in Ulu_Glass_averaged_renamed
Ulu_Glass_averaged_renamed = Ulu_Glass_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample_ID' else 'Sample')
Ulu_Glass_averaged_renamed['Sample'] = Ulu_Glass_averaged_renamed['Sample'].astype('float64')

# Merge DataFrames on 'Sample'
merged_df = combined_df_wglass.merge(Ulu_Glass_averaged_renamed, on='Sample', how='left', suffixes=('', '_new'))

# Update values in combined_df_wglass for matching rows
for column in Ulu_Glass_averaged_renamed.columns:
    if column != 'Sample':  # Skip the 'Sample' column as it's used for merging
        new_column = column + '_new'
        if new_column in merged_df.columns:
            # Update only non-NA values from the merged columns
            combined_df_wglass[column] = merged_df[new_column].combine_first(combined_df_wglass[column])

# Export the result to an Excel file
combined_df_wglass.to_excel(compilation_folder + slash + 'SWRZ_Raman_EDS_Gl_Merged.xlsx')
combined_df_wglass
```
