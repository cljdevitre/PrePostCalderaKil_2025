## Now let's merge the glass data for K24 SWRZ first, then Ulu data (1974) from Wieser et al. 

### First import the glass EDS from June 2024 eruption

```python
K20GlassEDS=pd.read_excel(MasterFolder+slash+'Data'+slash+'Kilauea2020-2023_SamplesandChemistryDataRelease_Revision1_08272024.xlsx',sheet_name='Glass EPMA Data',skiprows=2)
K20GlassEDS=K20GlassEDS.loc[:, ~K20GlassEDS.columns.str.contains('Unnamed')]
K20GlassEDS=K20GlassEDS.rename(columns={"Sample":"Sample_original"}).dropna(how='all').drop(K20GlassEDS.tail(1).index)
K20GlassEDS=K20GlassEDS[K20GlassEDS['Sample_original'].str.contains("KS20-527")]
K20GlassEDS['Sample'] = ['K20' for sample_name in K20GlassEDS['Sample_original']]
K20GlassEDS.insert(1, 'Sample', K20GlassEDS.pop('Sample'))

K20GlassEDS
```

### Average the Glass data

```python
grouped = K20GlassEDS.groupby('Sample')

# Separate numeric and non-numeric columns
numeric_cols = K20GlassEDS.select_dtypes(include='number')
non_numeric_cols = K20GlassEDS.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_K20GlassEDS_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_K20GlassEDS_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_K20GlassEDS_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate K20GlassEDS_averageds and reindex columns to the original order
K20GlassEDS_averaged = pd.concat([numeric_K20GlassEDS_averaged_mean, non_numeric_K20GlassEDS_averaged], axis=1)
K20GlassEDS_averaged = K20GlassEDS_averaged.reindex(columns=K20GlassEDS.columns)

# Add standard deviation columns after reindexing
K20GlassEDS_averaged = pd.concat([K20GlassEDS_averaged, numeric_K20GlassEDS_averaged_std.add_suffix('_STD')], axis=1)

# Reset index to get a DataFrame
K20GlassEDS_averaged = K20GlassEDS_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
K20GlassEDS_averaged = pd.merge(K20GlassEDS_averaged, row_counts[['Sample', 'averaged?']], on='Sample', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
K20GlassEDS_averaged['averaged?'] = K20GlassEDS_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# K20GlassEDS_averaged=K20GlassEDS_averaged[~(K20GlassEDS_averaged['filename'].str.contains("186-9-118-A"))]
# K20GlassEDS_averaged=K20GlassEDS_averaged.drop(columns={'Sample'})
K20GlassEDS_averaged_renamed = K20GlassEDS_averaged.rename(columns=lambda x: x + '_Gl' if x != 'Sample' else 'Sample')

K20GlassEDS_averaged_renamed.to_excel(compilation_folder+'/'+"SWRZ_GlassEDS_averaged.xlsx")


K20GlassEDS_averaged_renamed
```

### Combine the averaged glasses together

```python
averaged_glasses=K20GlassEDS_averaged_renamed.copy()
averaged_glasses
```

## Now merge and export the merged data

```python
combined_df['Sample'] = combined_df['Sample'].astype(str)
combined_df_wglass = combined_df.merge(averaged_glasses, on='Sample', how='left')
# combined_df_wglass.to_clipboard(excel=True)

combined_df_wglass.to_excel(compilation_folder+slash+'K20_Raman_EDS_Gl_Merged.xlsx')

combined_df_wglass
```

# Now let's merge the EBSD

```python
EBSD_data=pd.read_excel(MasterFolder +slash+"Data"+slash+'Data_KS20-527'+slash+'EBSD'+slash+'Combined_EBSD'+slash+'EBSD_combined_all.xlsx')
EBSD_data=EBSD_data.rename(columns={'EBSD_name':'EBSD_simple_name','grainID':'EBSD_grainID','Filename':'EBSD_filename'})

final_df=pd.merge(left=combined_df_wglass,on=['EBSD_simple_name','EBSD_grainID'],right=EBSD_data,suffixes=('', '_EBSDdata'),how='left')

##### Export fully merge data ####

final_df.to_excel(compilation_folder+slash+"KS20-527_Final_data_wEBSD.xlsx")
final_df
```
