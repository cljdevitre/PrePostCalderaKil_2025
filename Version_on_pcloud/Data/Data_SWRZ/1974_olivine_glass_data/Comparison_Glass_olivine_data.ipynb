```python
# These are olivines with melt inclusions from my Melt inclusion Nat Comms paper
Ulu_input=pt.import_excel('UluData.xlsx',
                         sheet_name='Melt Inclusions+host olivines')
Ulu_all=Ulu_input['my_input'] ## All columns
Ulu_Ols_all=Ulu_input['Ols'] ## Just Liquid columns
Ulu_Ols_19=Ulu_Ols_all.loc[Ulu_all['Eruption']==919]
Ulu_Ols_19.head()
```

```python
Ulu2_input=pt.import_excel('UluEBSDpaperData.xlsx',
                         sheet_name='KL0919  Olivine')
Ulu2_all=Ulu2_input['my_input'] ## All columns
Ulu_Ols_19_2=Ulu2_input['Ols'] ## Just Liquid columns


```

```python
combined_ols_19 = pd.concat([Ulu_Ols_19_2, Ulu_Ols_19], ignore_index=True)
```

```python
Ol_Fo_Calc_19=pt.calculate_ol_fo(ol_comps=combined_ols_19)
```

```python
Ulu_MG_input=pt.import_excel('UluData.xlsx',
                         sheet_name='Matrix Glass')
Ulu_all=Ulu_MG_input['my_input'] ## All columns
Ulu_Liqs_all=Ulu_MG_input['Liqs'] ## Just Liquid columns
Ulu_Liqs_19=Ulu_Liqs_all.loc[Ulu_all['Eruption']==919]
Ulu_Liqs_19.head()
```

```python
Liq_Mgno_calc=pt.calculate_liq_mgno(liq_comps=Ulu_Liqs_19, Fe3Fet_Liq=0.18)
Liq_Mgno_calc.head()
# Calculate mean value to plot
Liq_Mgno_calc_mean=np.mean(Liq_Mgno_calc)
```

```python
Rhodes=pt.calculate_ol_rhodes_diagram_lines(Min_Mgno=0.5, Max_Mgno=0.7)
Rhodes.head()
```

## Lets add new data

```python
# Olivines
Ol_new=pd.read_excel(r'../EDS_olivine.xlsx')
Ol_new_2024=Ol_new.loc[Ol_new['Sample Name'].str.contains('K24')]
Ol_new_1974=Ol_new.loc[Ol_new['Sample Name'].str.contains('919')]
```

```python
## Glasses 
glass=pd.read_excel(r'../GlassData2024.xlsx')
glass=glass.loc[~glass['Sample Name'].str.contains('JDF')]
Liq_Mgno_calc_mean_2024=np.mean(glass['Mg#'])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

# Plotting for Roeder and Emslie
ax1.set_title('Roeder and Emslie, 1970')
# Plotting equilibrium lines
ax1.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Roeder (Kd=0.27)'], ':k')
ax1.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Roeder (Kd=0.33)'], ':k')
ax1.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Roeder (Kd=0.3)'], '-k')

# Plotting data
ax1.plot(Ol_Fo_Calc_19*0+Liq_Mgno_calc_mean, Ol_Fo_Calc_19, 'ok', mfc='red', ms=3, label='Cam17 - 1974')
ax1.plot(Ol_new_1974['Fo']*0+Liq_Mgno_calc_mean, Ol_new_1974['Fo'], 'xr', mfc='red', label='UCB24 - 1974')
ax2.plot(Ol_new_1974['Fo']*0+Liq_Mgno_calc_mean, Ol_new_1974['Fo'], 'xr', mfc='red')

ax1.plot(Ol_new_2024['Fo']*0+Liq_Mgno_calc_mean_2024, Ol_new_2024['Fo'], 'xc', mfc='c', label='UCB24 - 2024')
ax2.plot(Ol_new_2024['Fo']*0+Liq_Mgno_calc_mean_2024, Ol_new_2024['Fo'], 'xc', mfc='c')

ax2.set_title('Shea (2024)')
# Plotting equilibrium lines
ax2.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Shea (Kd=0.325)'], ':k')
ax2.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Shea (Kd=0.335)'], '-k')
ax2.plot(Rhodes['Mg#_Liq'], Rhodes['Eq_Ol_Fo_Shea (Kd=0.345)'], ':k')
# Plotting data
ax2.plot(Ol_Fo_Calc_19*0+Liq_Mgno_calc_mean, Ol_Fo_Calc_19, 'ok', mfc='red')
ax1.set_ylabel('Ol Fo content')
ax1.set_xlabel('Glass Mg#')
ax2.set_xlabel('Glass Mg#')
ax1.set_ylim([0.74, 0.92])
ax2.set_ylim([0.74, 0.92])
ax1.set_xlim([0.5, 0.7])
ax2.set_xlim([0.5, 0.7])
ax1.legend()
```
